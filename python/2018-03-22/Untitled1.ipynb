```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap = cv2.VideoCapture(0)

while(True):
    ret, frame = cap.read()
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)

    cv2.imshow('frame', rgb)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        out = cv2.imwrite('capture.jpg', frame)
        break

cap.release()
cv2.destroyAllWindows()
```

```python
try:
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)
    rgb = np.flip(rgb[:,:,0:3], axis=2)
    plt.imshow(rgb)
    mask = np.zeros(rgb.shape, dtype=np.uint8)
    mask[300:700,300:700,:] = 1;
    plt.figure()
    plt.imshow(rgb*mask)

    plt.figure()
    plt.title("Color Histogram")
    plt.xlabel("Bins")
    plt.ylabel("# of Pixels")
    colors = "rgb"
    for (chan, color) in zip(rgb, colors):
        hist = cv2.calcHist([chan[300:700,300:700]], [0], None, [256], [0, 256])
        plt.plot(hist, color=color)
        plt.xlim([0, 256])
except Exception as e:
    print(e)
finally:
    cap.release()
    cv2.destroyAllWindows()

```
